```python
data = {
    'role' : [], #text
    'company' : [], #direct
    'jd' : [], # from lists
    'requirements' : [], # from lists
    'lifeAt' : [], # additionalPlain
    'commitment' : [], #categories
    'department' : [], #categories
    'location' : [], # categories
    'team': [], # Categories
    'createdAt' : [],
    'companyDescription': [],
    'id': [],
    'country': [],
    'type': [],
    'hostedUrl': [],
    'applyUrl': [],
}

for name in files:
    with open(os.path.join(base, name), 'r', encoding="utf8") as f:
        jobs = json.load(f)
        for j in jobs:
            data['role'].append(j['text'] if 'text' in j.keys() else "")
            data['company'].append(name.split('.')[0])
            data['lifeAt'].append(j['additionalPlain'] if 'additionalPlain' in j.keys() else "")
            data['commitment'].append(j['categories']['commitment'] if ('categories' in j.keys()) and ('commitment' in j['categories'].keys()) else "")
            data['department'].append(j['categories']['department'] if ('categories' in j.keys()) and ('department' in j['categories'].keys()) else "")
            data['location'].append(j['categories']['location'] if ('categories' in j.keys()) and ('department' in j['categories'].keys()) else "")
            data['team'].append(j['categories']['team'] if ('categories' in j.keys()) and ('team' in j['categories'].keys()) else "")
            data['id'].append(j['id'])
            data['createdAt'].append(j['createdAt'] if 'createdAt' in j.keys() else "")
            data['companyDescription'].append(j['descriptionPlain'] if 'descriptionPlain' in j.keys() else "")
            data['country'].append(j['country'] if 'country' in j.keys() else "")
            data['type'].append(j['workplaceType'] if 'workplaceType' in j.keys() else "")
            data['hostedUrl'].append(j['hostedUrl'] if 'hostedURL' in j.keys() else "")
            data['applyUrl'].append(j['applyUrl'] if 'applyURL' in j.keys() else "")

            if 'lists' in j.keys():
                data['jd'].append(j['lists'][0]['content'] if len(j['lists']) == 1 else "")
                data['requirements'].append(j['lists'][1]['content'] if len(j['lists']) == 2 else "")
            else:
                print(j)
                break
                data['jd'].append("")
                data['requirements'].append("")
```

```python
df = pd.DataFrame.from_dict(data)
df.to_csv('main.csv', index=False)
```
