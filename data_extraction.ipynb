```python
files = ['cred', 'Meesho']
base = 'data'
```

```python
data = {
    # 'role': [],
    'company' : [],
    'title' : [],
    'jd' : [],
    # 'teamDescription' : [],
    'requirements' : [],
}
for name in files:
    with open(os.path.join(base, f'{name}.json'), 'r', encoding="utf8") as f:
        jobs = json.load(f)
        for i, j in enumerate(jobs):
            data['company'].append(name)
            data['title'].append(j['text'])
            
            if len(j['lists']) == 2:
                data['jd'].append(j['lists'][0]['content'])
                data['requirements'].append(j['lists'][1]['content'])
            else:
                data['jd'].append("")
                data['requirements'].append("")
```

```python
import pandas as pd
pd.DataFrame.from_dict(data)
```
